```python
%matplotlib inline
import networkx as nx
import matplotlib.pylab as plt
```

```python
nodes = [
    "answer",
    "assignment",
    "course",
    "event_course_activity",
    "handin",
    "institution",
    "question",
    "question_section",
    "report_grade",
    "student_evaluation",
    "user"
]

edges = [
    ("answer", "question"),
    ("answer", "report_grade"),
    ("assignment", "question_section"),
    ("assignment", "user"),
    ("assignment", "course"),
    ("course", "user"),
    ("course", "institution"),
    ("event_course_activity", "course"),
    ("event_course_activity", "user"),
    ("event_course_activity", "handin"),
    ("question_section", "question"),
    ("report_grade", "user"),
    ("report_grade", "handin"),
    ("report_grade", "assignment"),
    ("student_evaluation", "user"),
    ("student_evaluation", "course")
]
```

```python
G = nx.DiGraph()
```

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
nx.draw(G)
```
